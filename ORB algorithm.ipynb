```python
import cv2
import matplotlib.pyplot as plt


img1=cv2.imread(r"C:\Users\zbook 17 g3\Desktop\New folder (4)\Screenshot 2023-07-24 184457.png",0)
img2=cv2.imread(r"C:\Users\zbook 17 g3\Desktop\New folder\Screenshot 2023-07-24 184457.png",0)

orb=cv2.ORB_create()


keypiont1,describe1=orb.detectAndCompute(img1,None)
keypiont2,describe2=orb.detectAndCompute(img2,None)

matcherr=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=False)


matches=matcherr.match(describe1,describe2)
matches=sorted(matches,key=lambda x: x.distance)


matches_show=cv2.drawMatches(img1,keypiont1,img2,keypiont2,matches[:8],None,flags=2)

plt.figure(figsize=(20,20))
plt.axis("off")
plt.title("ORB")
plt.imshow(matches_show)
```
